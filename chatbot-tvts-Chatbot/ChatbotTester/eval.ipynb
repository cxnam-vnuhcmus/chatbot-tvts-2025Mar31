```python
from ChatbotTester.bot import get_chatbot_evaluation_instance


chat = get_chatbot_evaluation_instance()

def evaluate(session_id="", chunk_id=""):
  gen = chat.ask(session_id=session_id, chunk_id=chunk_id)
  while True:
    try:
      cmd, msg = next(gen)
      print(f"{cmd}: {msg}")
    except StopIteration:
      break
  print("======================")
```
